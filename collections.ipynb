### Tableaux en Kotlin

```kotlin
val rockPlanets = arrayOf<String>("Mercury", "Venus", "Earth", "Mars")
val gasPlanets = arrayOf("Jupiter", "Saturn", "Uranus", "Neptune")
val solarSystem = rockPlanets + gasPlanets
println(solarSystem[0])
println(solarSystem[1])
println(solarSystem[2])
println(solarSystem[3])
println(solarSystem[4])
println(solarSystem[5])
println(solarSystem[6])
println(solarSystem[7])
```

### Listes

```kotlin
val solarSystem = listOf("Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune")
println(solarSystem.size)
println(solarSystem[2])
println(solarSystem.get(3))
println(solarSystem.indexOf("Earth"))
println(solarSystem.indexOf("Pluto")) //-1
```

### Itérer des éléments de liste à l'aide d'une boucle for

```kotlin
for (planet in solarSystem) {
    println(planet)
}
```

### Ajouter des éléments à une liste

```kotlin
val solarSystem = mutableListOf("Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune")
solarSystem.add("Pluto")
solarSystem.add(3, "Theia")
solarSystem[3] = "Future Moon"
println(solarSystem[3])
println(solarSystem[9])
```

### Supprimer des éléments d'une liste

```kotlin
val solarSystem = mutableListOf("Mercury", "Venus", "Earth", "Future Moon", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune", "Pluto")
println(solarSystem.size)
solarSystem.removeAt(9)
println(solarSystem.size)
solarSystem.remove("Future Moon")
println("contains Pluto: ${solarSystem.contains("Pluto")}")
println("Future Moon" in solarSystem)
```

### Set

```kotlin
val solarSystem = mutableSetOf("Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune")
println(solarSystem.size)
solarSystem.add("Pluto")
println(solarSystem.size)
println(solarSystem.contains("Pluto"))
solarSystem.remove("Pluto")
println(solarSystem.contains("Pluto"))
```

### Map

```kotlin
val solarSystem = mutableMapOf(
    "Mercury" to 0,
    "Venus" to 0,
    "Earth" to 1,
    "Mars" to 2,
    "Jupiter" to 79,
    "Saturn" to 82,
    "Uranus" to 27,
    "Neptune" to 14
)
println(solarSystem.size)
solarSystem["Pluto"] = 5
println(solarSystem.size)
```
